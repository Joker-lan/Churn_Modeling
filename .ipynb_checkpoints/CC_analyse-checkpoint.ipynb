```python
def yule_skewness_quartiles(data):
    # Vérifier que les données ne sont pas vides
    if len(data) == 0:
        return np.nan
    
    # Calculer les quartiles et la médiane
    Q1 = np.percentile(data, 25)
    Q2 = np.median(data)  # Médiane
    Q3 = np.percentile(data, 75)

    # Calculer le coefficient d'asymétrie de Yule
    numerator = (Q3 - Q2) - (Q2 - Q1)
    denominator = (Q3 - Q2) + (Q2 - Q1)

    # Éviter la division par zéro
    if denominator == 0:
        return np.nan
    
    skewness = numerator / denominator
    
    return skewness

# Exemple d'utilisation avec une colonne de votre DataFrame
# Par exemple, pour la variable 'Balance'
yule_skewness_balance = yule_skewness_quartiles(churn_modeling['CreditScore'])
print(f"Le coefficient d'asymétrie de Yule pour la variable CreditScore est : {yule_skewness_balance:.4f}")
```

```python
#TESTS POUR LE SCORE AVEC EXITED

quit = churn_modeling[churn_modeling["Exited"] == 1] 
non_quit = churn_modeling[churn_modeling["Exited"] == 0]

quit = quit[quit["CreditScore"] >= 0]
non_quit = non_quit[non_quit["CreditScore"] >= 0]

shapiro_results_quit = shapiro(quit["CreditScore"])
shapiro_results_non_quit = shapiro(non_quit["CreditScore"])
print("Résultats du test de Shapiro pour ceux qui sont partis :", shapiro_results_quit)
print("Résultats du test de Shapiro pour ceux qui sont encore là :", shapiro_results_non_quit)
```

```python
def mannwhiteney_test(data, var, target='Exited', alpha=0.05):
    positive = data[data[target] == 1]
    negative = data[data[target] == 0]
    
    if positive.shape[0] > negative.shape[0]:
        positive = positive.sample(negative.shape[0])
    else:
        negative = negative.sample(positive.shape[0])
        
    stat, p_value = mannwhitneyu(negative[var].dropna(), positive[var].dropna())
    
    if p_value < alpha:
        message = "L'hypothèse H0 est rejetée !" 
    else:
        message = "L'on ne peut conclure au rejet de l'hypothèse H0 !"
    
    return stat, p_value, message

stat, p_value, message = mannwhiteney_test(churn_modeling, "CreditScore")
print(p_value)
```

```python
def mannwhiteney_test(data, var, target='Exited', alpha=0.05):
    positive = data[data[target] == 1]
    negative = data[data[target] == 0]
    
    if positive.shape[0] > negative.shape[0]:
        positive = positive.sample(negative.shape[0])
    else:
        negative = negative.sample(positive.shape[0])
        
    stat, p_value = mannwhitneyu(negative[var].dropna(), positive[var].dropna())
    
    if p_value < alpha:
        message = "L'hypothèse H0 est rejetée !" 
    else:
        message = "L'on ne peut conclure au rejet de l'hypothèse H0 !"
    
    return stat, p_value, message

stat, p_value, message = mannwhiteney_test(churn_modeling, "Balance")
print(p_value)
```

```python
np.random.seed(0)  # Pour la reproductibilité
churn_modeling = pd.DataFrame({
    'Balance': np.random.gamma(shape=2, scale=2000, size=10000)  # Exemple de distribution
})

# 1. Calculer l'Indice de Gini
def gini_coefficient(data):
    n = len(data)
    if n == 0:
        return 0
    # Tri des données
    sorted_data = np.sort(data)
    cumulative_values = np.cumsum(sorted_data)
    # Calcul de l'aire sous la courbe de Lorenz
    lorenz_curve = cumulative_values / cumulative_values[-1]
    gini = 1 - 2 * np.trapz(lorenz_curve, dx=1/n)
    return gini

# Calculer l'indice de Gini pour la variable Balance
gini_balance = gini_coefficient(churn_modeling['Balance'])
print(f"Indice de Gini pour la variable Balance : {gini_balance:.4f}")

# 2. Tracer la Courbe de Lorenz
def plot_lorenz_curve(data):
    sorted_data = np.sort(data)
    cumulative_values = np.cumsum(sorted_data)
    lorenz_curve = cumulative_values / cumulative_values[-1]

    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 1, len(lorenz_curve)), lorenz_curve, label='Courbe de Lorenz', color='blue')
    plt.plot([0, 1], [0, 1], label='Égalité parfaite', color='red', linestyle='--')
    plt.fill_between(np.linspace(0, 1, len(lorenz_curve)), lorenz_curve, color='lightblue', alpha=0.3)
    plt.title('Courbe de Lorenz pour la variable Balance')
    plt.xlabel('Fraction cumulée de la population')
    plt.ylabel('Fraction cumulée de la Balance')
    plt.legend()
    plt.grid()
    plt.savefig('lorentz_balance.png')
    plt.show()
    
# Tracer la courbe de Lorenz
plot_lorenz_curve(churn_modeling['Balance'])
```

```python
# Fonction pour calculer l'indice de Gini
def gini_coefficient(data):
    data = np.array(data)
    
    n = data.size
    if n == 0 or n == 1:
        return 0
    
    data_sorted = np.sort(data)
    cumulative_values = np.cumsum(data_sorted)
    
    gini = (2 * np.sum(cumulative_values) / cumulative_values[-1] - (n + 1)) / n
    
    return gini

gini_index_estimated_salary = gini_coefficient(churn_modeling['EstimatedSalary'])
print(f"L'indice de Gini pour la variable EstimatedSalary est : {gini_index_estimated_salary:.4f}")
```
