```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import chi2_contingency, ttest_ind, shapiro
from scipy.stats import mannwhitneyu, kruskal, wilcoxon
```

```python
churn_modeling = pd.read_csv("Churn_Modelling.csv", sep = ",", decimal = ".")
```

```python
churn_modeling.head(10)
```

```python
churn_modeling["Exited"].value_counts()
```

```python
freq = churn_modeling["Exited"].value_counts(normalize = True)
plt.figure(figsize = (10,10))
plt.pie(freq,
       labels = ["Présent à la banque", "Ont quitté la banque"],
       colors = ["turquoise", "salmon"],
       explode = [0, 0.05],
       autopct = lambda x: str(np.round(x, 2)) + "%",
       shadow = True)
plt.legend()
plt.savefig('figure_exited.png')
plt.show() # pour ne plus afficher le texte embetant
```

```python
mean_solde_quit = churn_modeling[churn_modeling['Exited'] == 1]['Balance'].mean()
```

```python
mean_solde_quit
```

```python
mean_solde_not_quit = churn_modeling[churn_modeling['Exited'] == 0]['Balance'].mean()
```

```python
mean_solde_not_quit
```

```python
churn_modeling["Geography"].value_counts()
```

```python
frequencies = churn_modeling['Geography'].value_counts()

plt.figure(figsize=(6, 6))
frequencies.plot(kind='bar', color='skyblue')

# Ajout des titres et des labels
plt.title('Localisation des individus par pays', fontsize=16)
plt.xlabel('Pays', fontsize=14)
plt.ylabel('Nombre d\'individus de la banque par pays', fontsize=14)
plt.xticks(rotation=45)  # Rotation des labels sur l'axe des x

plt.savefig('figure_localisation.png')
plt.show()
```

```python
frequencies = churn_modeling.groupby(['Geography', 'Exited']).size().unstack()

percentages = frequencies.div(frequencies.sum(axis=1), axis=0) * 100

ax = percentages.plot(kind='bar', stacked=True, color=['turquoise', 'salmon'], figsize=(10, 6))

plt.title('Proportions d\'individus selon l\'état (Exited) par localisation', fontsize=16)
plt.xlabel('Localisation', fontsize=14)
plt.ylabel('Pourcentage d\'individus', fontsize=14)
plt.xticks(rotation=45)

plt.legend(title='État', labels=['Restés', 'Quités'])

for i in range(percentages.shape[0]):
    for j in range(percentages.shape[1]):
        percentage = percentages.iloc[i, j]
        ax.text(i, percentage/2 + percentages.iloc[i, :j].sum(), f'{percentage:.1f}%', ha='center', va='center', color='black')
plt.tight_layout()
plt.savefig('figure_exitedVSlocalisation.png')
plt.show()
```

```python
tabI = pd.crosstab(churn_modeling["Exited"], churn_modeling["Geography"])
print(tabI)
```

```python
chi2_stat, p_value, ddl, effectifs_theoriques = chi2_contingency(tabI)
```

```python
p_value
```

```python
frequencies = churn_modeling['Gender'].value_counts()

plt.figure(figsize=(6, 6))
frequencies.plot(kind='bar', color='skyblue')

# Ajout des titres et des labels
plt.title('Sexe des individus', fontsize=16)
plt.xlabel('Sexe', fontsize=14)
plt.ylabel('Nombre d\'individus de la banque par Sexe', fontsize=14)
plt.xticks(rotation=45)

plt.savefig('figure_sexe.png')
plt.show()
```

```python
frequencies = churn_modeling.groupby(['Gender', 'Exited']).size().unstack()

# Conversion des fréquences en pourcentages
percentages = frequencies.div(frequencies.sum(axis=1), axis=0) * 100

plt.figure(figsize=(10, 6))  
ax = percentages.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])

plt.title('Proportions d\'individus selon l\'état (Exited) par sexe', fontsize=14)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Pourcentage d\'individus', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotation et alignement des étiquettes

plt.legend(title='État', labels=['Restés', 'Quités'])

for i in range(frequencies.shape[0]):
    for j in range(frequencies.shape[1]):
        count = frequencies.iloc[i, j]
        ax.text(i, count/2 + frequencies.iloc[i, :j].sum(), f'{count}', 
                ha='center', va='center', color='black')

plt.subplots_adjust(bottom=0.2, top=0.9, left=0.1, right=0.9)

plt.tight_layout()
plt.savefig('figure_exitedVSsexe.png')
plt.show()
```

```python
tabII = pd.crosstab(churn_modeling["Exited"], churn_modeling["Gender"])
print(tabII)
```
